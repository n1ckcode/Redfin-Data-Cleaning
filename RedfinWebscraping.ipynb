# Structure data and export into CSV file

```python
import requests
from bs4 import BeautifulSoup
import csv

headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:108.0) Gecko/20100101 Firefox/108.0"}

data = []

for i in range(1, 10):
    url = f"https://www.redfin.com/state/California/apartments-for-rent/filter/viewport=38.13751:37.17207:-121.37222:-123.06412/page-{i}"
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')

    prices = [item.get_text() for item in soup.find_all(class_='homecardV2Price')]
    addresses = [item.get_text() for item in soup.find_all(class_='homeAddressV2')]
    homestats = [item.get_text() for item in soup.find_all(class_='HomeStatsV2 font-size-small')]

    for j in range(len(prices)):
        data.append([prices[j], addresses[j], homestats[j]])

with open("apartment_data.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Price", "Address", "Home Stats"])
    writer.writerows(data)



print("created")
```

# Data cleaning and restructuring 

```python
# remove the '|'' and every string before it from values in the "Address" column
dff['Address'] = dff['Address'].str.split('|').str[-1]


dff


```

```python
# split the values in the "Address" column of the DataFrame dff into separate columns for "Address", "City", and "State".
dff['Address'].str.rsplit(',', 2).apply(lambda x: x[0] if len(x) == 3 else x[0] + ',' + x[1])
dff = dff.assign(
    Address=dff['Address'].str.rsplit(',', 2).apply(lambda x: x[0] if len(x) == 3 else x[0] + ',' + x[1]),
    City=dff['Address'].str.rsplit(',', 2).apply(lambda x: x[1] if len(x) == 3 else None),
    State=dff['Address'].str.rsplit(',', 2).apply(lambda x: x[2] if len(x) == 3 else None)
)
dff
```

```python
# extract the ZipCode from the "State" column of the DataFrame dff and then remove it from the "State" column, creates a new column for the zip code
dff['ZipCode'] = dff['State'].str.extract(r'(\d{5})')
dff['State'] = dff['State'].str.replace(r' \d{5}', '')

dff
```

```python
# replace values in the 'Home Stats' column of a Pandas DataFrame dff. The argument r'\d+-' is a regular expression pattern that matches one or more consecutive digits (\d+) followed by a hyphen (-).

dff['Home Stats'] = dff['Home Stats'].str.replace(r'\d+-', '')

dff
```

```python
# change the name of a column in the dff DataFrame.
dff.rename(columns={"Home Stats": "Size"}, inplace=True)
dff
```

```python
# re format returns a formatted string based on the contents of s

import re 
def format_size(s):
    parts = re.findall(r'(\d+)\s*[Bb]ed.*?(\d+)\s*[Bb]ath.*?(\d+,\d+)\s*[Ss]q\s*[.]\s*[Ff]t\.', s)
    if parts:
        beds, baths, sqft = parts[0]
        return f"{beds} Beds, {baths} Baths, {sqft} Sq. Ft."
    return s

dff["Size"] = dff["Size"].apply(format_size)


dff
```

```python
# reformat the values in the "Size" column by using string operations and regular expressions.

import re

def format_size(s):
    parts = re.findall(r'(\d+)\s*[Bb]ed.*?(\d+)\s*[Bb]ath.*?(\d+,\d+)\s*[Ss]q\s*[.]\s*[Ff]t\.', s)
    if parts:
        beds, baths, sqft = parts[0]
        if beds == '1':
            beds = '1 Bed'
        else:
            beds = f"{beds} Beds"
        if baths == '1':
            baths = '1 Bath'
        else:
            baths = f"{baths} Baths"
        return f"{beds}, {baths}, {int(sqft.replace(',', '')):,} Sq. Ft."
    parts = re.findall(r'(\d+)\s*[Bb]ed.*?(\d+)\s*[Bb]ath.*?(\d+)\s*[Ss]q\s*[.]\s*[Ff]t\.', s)
    if parts:
        beds, baths, sqft = parts[0]
        if beds == '1':
            beds = '1 Bed'
        else:
            beds = f"{beds} Beds"
        if baths == '1':
            baths = '1 Bath'
        else:
            baths = f"{baths} Baths"
        return f"{beds}, {baths}, {int(sqft):,} Sq. Ft."
    return s

dff["Size"] = dff["Size"].apply(format_size)

dff
```

```python

# first we need to remove the "/mo" part of the string
dff['Price'] = dff['Price'].str.replace("/mo", "")

# next, we need to remove the dash and everything after it
dff['Price'] = dff['Price'].str.replace(r'-.*', '')

# finally, we need to remove the comma separators in the numbers and add a "$" sign
dff['Price'] = "$" + dff['Price'].str.replace(',', '')

dff
```

```python
# remove the extra "$" sign
dff['Price'] = dff['Price'].str.replace("$", "")
# add the "$" sign
dff['Price'] = "$" + dff['Price']

dff

```

```python
# change the name of the "Price" column to "Monthly Rent"
dff.rename(columns={'Price': 'Monthly Rent'}, inplace=True)

dff
```

```python
# Extract the values using regex
dff[['Beds', 'Baths', 'SqFt']] = dff['Size'].str.extract(r'(\d+) Beds, (\d+) Baths, (\d+,\d+) Sq. Ft.', expand=True)

# Replace NaN values with 0
dff['SqFt'] = dff['SqFt'].str.replace(',', '').astype(float).fillna(0).astype(int)


dff




```

```python
# drop the size column
dff.drop(['Size'], axis=1, inplace=True)

dff
```

```python

# Replace NaN values with 0
dff['Beds'].fillna(0, inplace=True)
dff['Baths'].fillna(0, inplace=True)

dff
```
